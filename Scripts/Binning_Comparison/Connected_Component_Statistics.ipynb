```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from os import listdir

filepath_repeats_present = '/Users/harihara/Mount/Hot_Springs_Analysis/Connected_Components_Wo_Repeats/'
filepath_repeats_removed = '/Users/harihara/Mount/Hot_Springs_Analysis/Connected_Components/'
files = listdir(filepath_repeats_present)

op_list = []

for f in files:
    print(f)
    if f[0] == '.':
        continue
    sample_id = f.replace("_scaffolds_no_rd.csv","")
    try:
        df_repeats_present = pd.read_csv(filepath_repeats_present+f)
    except FileNotFoundError:
        continue
    df_repeats_present['Bin'].fillna(-1, inplace = True)
    binned_length = df_repeats_present[df_repeats_present['Bin'] > 0]['Length'].sum()
    unbinned_length = df_repeats_present[df_repeats_present['Bin'] <= 0]['Length'].sum()
    unbinned_gr_3000 = len(df_repeats_present[(df_repeats_present['Bin'] <= 0) & 
                                              (df_repeats_present['Length'] >= 3000)])
    unbinned_lt_3000 = len(df_repeats_present[(df_repeats_present['Bin'] <= 0) & 
                                              (df_repeats_present['Length'] < 3000)])
    d = {'Sample':sample_id, 
         '#Bins':len(df_repeats_present['Bin'].unique()),
         '#Connected_Components':len(df_repeats_present['Connected_Component'].unique()),
         '#Contigs':len(df_repeats_present),
         '#Binned_Contigs':len(df_repeats_present[df_repeats_present['Bin'] > 0]),
         '#Contigs_Bin=0' : len(df_repeats_present[df_repeats_present['Bin'] == 0]),
         '#Contigs_Bin=-1' : len(df_repeats_present[df_repeats_present['Bin'] == -1]),
         'Binned Length':binned_length,
         'Unbinned Length':unbinned_length,
         'Unbinned Length (>3kb)':unbinned_gr_3000,
         'Unbinned Length (<3kb)':unbinned_lt_3000
        }
    op_list.append(d)
df_op = pd.DataFrame(op_list).set_index('Sample')
```

```python
df_op.to_csv('/Users/harihara/Mount/Hot_Springs_Analysis/Connected_Component_Analysis.csv')
```
