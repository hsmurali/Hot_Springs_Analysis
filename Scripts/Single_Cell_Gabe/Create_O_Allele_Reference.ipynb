```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Mount/hotspring_metagenome/single_cell_analysis_with_Gabe_Birzu/recruitment_v2/metagenome_recruitment_target_sites.tsv', sep = "\t")
df_o_group = df[df['notes'] == 'O orthogroup cluster alleles']
genes = df_o_group['Unnamed: 0'].tolist()
```

```python
total_seqs = 0
o = open("O_Orthogroup.fna","w")
for g in genes:
    r = open('Mount/hotspring_metagenome/single_cell_analysis_with_Gabe_Birzu/recruitment_v2/'+g+'.fna').read()
    o.write(r)
    print(g,r.count(">"))
    total_seqs += r.count(">")
o.close()
```
