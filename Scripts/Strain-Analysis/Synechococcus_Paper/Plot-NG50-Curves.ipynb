```python
import numpy as np
import pandas as pd
from os import listdir
import matplotlib.pyplot as plt

rcParams = {'font.size': 30, 'font.weight': 'normal', 'font.family': 'sans-serif',
            'axes.unicode_minus':False, 'axes.labelweight':'normal'}

plt.rcParams.update(rcParams)
```

```python
data_dir = '/Users/harihara/Mount-2/hotspring_metagenome/osa_osb_reference_based_contigs/osa_final_contigs/'
files = listdir(data_dir)
osa = {}

for f in files:
    if f.endswith(".fai"):
        df = pd.read_csv(data_dir+f, sep = "\t", names = ['Seq','Len','1','2','3','4'])
        df = df.sort_values(by = ['Len'], ascending = False)
        osa[f.replace("_FD_osa_contigs.fna.fai","")] = df['Len'].tolist()
        
fig,ax = plt.subplots(1,1,figsize = (16,16))
for o in osa:
    ax.plot(osa[o], np.cumsum(osa[o]),color = 'black')
ax.set_xlim(ax.get_xlim()[::-1])
ax.set_xlabel('Contigs sorted in decreasing orders of their lengths')
ax.set_ylabel('Cumulative number of basepairs assembled')
fig.tight_layout()
fig.savefig('/Users/harihara/Research-Activities/Plots/Hot_Spring_Plots/Synechococcus-Paper/OSA-NG50-Denovo.pdf')
```
