```python
from os.path import isdir
from datetime import datetime
from os import listdir, mkdir
from Scaffold_Using_References_Utils import *
```

```python
outdir = '/Users/harihara/Research-Activities/Data/Hot-Spring/Missing_Contig_Coords_2/'

if not isdir(outdir):
    mkdir(outdir)
if not isdir(outdir+'OSA/'):
    mkdir(outdir+'OSA/')
if not isdir(outdir+'OSB/'):
    mkdir(outdir+'OSB/')
```

```python
data_dir = '/Users/harihara/Mount-2/hotspring_metagenome/Synechococcus_paper_analysis/reassembly/'
files = listdir(data_dir)
files.sort()

for f in files:
    if "_megahit_assembled_contigs_" in f:
        if "osa" in f: 
            genome = "OSA"
            secondary = "OSB"
        elif "osb" in f: 
            genome = "OSB"
            secondary = "OSA"
        print(datetime.now(), f, genome, secondary)
        
        sample_id = f.replace("_megahit_assembled_contigs_"+genome.lower(),"")
        osa_path = data_dir+'contig_mapping_paf/'+sample_id+'_'+genome.lower()+'_contigs_aligned_to_osa.paf'
        osb_path = data_dir+'contig_mapping_paf/'+sample_id+'_'+genome.lower()+'_contigs_aligned_to_osb.paf'
        g_path =  data_dir+f+'/'+sample_id+'_'+genome.lower()+'_scaffolds/oriented.gml'
        df_osa = Load_PAF(osa_path)
        df_osb = Load_PAF(osb_path)
        
        if genome == "OSA":
            primary_alignments = df_osa
            secondary_alignments = df_osb
        if genome == "OSB":
            primary_alignments = df_osb
            secondary_alignments = df_osa
           
        G = nx.read_gml(g_path)
        G = Add_Backbone_Alignment_Information(primary_alignments, genome, G, 80)
        G = Add_Additional_Attributes(secondary_alignments, secondary, G, 80)
        
        connected_components = list(nx.weakly_connected_components(G))
        out_list = []
        for c in connected_components:
            subg = G.subgraph(c)
            out = Assign_Coordinates(subg, genome)
            out_list += out
            
        df_scaffolds = pd.DataFrame(out_list)
        df_scaffolds.to_csv(outdir+genome+'/'+sample_id+'.txt', sep = "\t")
        print(genome, sample_id, len(df_scaffolds), datetime.now(), "\n\n\n")
        
```
