```python
from os.path import isdir
from datetime import datetime
from os import listdir, mkdir
from Scaffold_Using_References_Utils import *
```

```python
outdir = '/Users/harihara/Research-Activities/Data/Hot-Spring/Missing_Contig_Coords_Aug/'

if not isdir(outdir):
    mkdir(outdir)
if not isdir(outdir+'OSA/'):
    mkdir(outdir+'OSA/')
if not isdir(outdir+'OSB/'):
    mkdir(outdir+'OSB/')
```

```python
data_dir = '/Users/harihara/Mount-2/hotspring_metagenome/Synechococcus_paper_analysis/reassembly/'
assembly_dir = data_dir+'assembly_and_scaffolds/'
contig_align_path = data_dir+'contig_mapping_paf/'
```

```python
####Scaffold OSA Reassembly
files = listdir(assembly_dir+'osa/')
files.sort()
for f in files:
    if(f.startswith("Hot")):
        genome, secondary = 'OSA', 'OSB'
        print(datetime.now(), f, genome, secondary)
        sample_id = f.replace("_megahit_assembled_contigs_"+genome.lower(),"")
        osa_path = contig_align_path+sample_id+'_'+genome.lower()+'_contigs_aligned_to_osa.paf'
        osb_path = contig_align_path+sample_id+'_'+genome.lower()+'_contigs_aligned_to_osb.paf'
        g_path =  assembly_dir+'osa/'+f+'/'+sample_id+'_'+genome.lower()+'_scaffolds/oriented.gml'
        primary_alignments = Load_PAF(osa_path)
        secondary_alignments = Load_PAF(osb_path)
        
        G = nx.read_gml(g_path)
        G = Add_Backbone_Alignment_Information(primary_alignments, genome, G, 80)
        G = Add_Additional_Attributes(secondary_alignments, secondary, G, 80)
        
        connected_components = list(nx.weakly_connected_components(G))
        out_list = []
        for c in connected_components:
            subg = G.subgraph(c)
            out = Assign_Coordinates(subg, genome)
            out_list += out
            
        df_scaffolds = pd.DataFrame(out_list)
        df_scaffolds.to_csv(outdir+genome+'/'+sample_id+'.txt', sep = "\t")
```
