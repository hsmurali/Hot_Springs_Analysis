```python
import numpy as np
import pandas as pd
from os import listdir
import matplotlib.pyplot as plt

rcParams = {'font.size': 30, 'font.weight': 'normal', 'font.family': 'sans-serif',
            'axes.unicode_minus':False, 'axes.labelweight':'normal'}

plt.rcParams.update(rcParams)
```

```python
df = pd.read_csv('/Users/harihara/Mount-2/Coverage_Statistics.txt', sep = "\t")
```

```python
plt.rcParams.update(rcParams)

fig,ax = plt.subplots(1,2,figsize = (16,10))
OSA = df[df['Genome'] == 'OSA']
OSB = df[df['Genome'] == 'OSB']

OSA.plot.scatter('Breadth','COV', color = 'red', ax = ax[0], s = OSA['Max_Len']*0.005)
OSB.plot.scatter('Breadth','COV', color = 'blue', ax = ax[1], s = OSB['Max_Len']*0.005)

ax[0].set_xlabel('Breadth of Coverage')
ax[1].set_xlabel('Breadth of Coverage')
ax[0].set_ylabel('Coefficient of Variation of Coverage')
ax[1].set_ylabel("")

ax[0].set_title('OSA')
ax[1].set_title('OSB')

ax[1].set_ylim(ax[0].get_ylim())

ax[0].axhline(1, color = 'black', linestyle = '--', linewidth = 2)
ax[1].axhline(1, color = 'black', linestyle = '--', linewidth = 2)

fig.tight_layout()

fig.savefig('/Users/harihara/Research-Activities/Plots/Hot_Spring_Plots/Synechococcus-Paper/Coverage-Stats.pdf')
```
