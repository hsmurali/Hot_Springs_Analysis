```python
outpath = '/Users/harihara/Research-Activities/Plots/Hot_Spring_Plots/Single_Cell_Analyses/'

plt.rcParams.update(rcParams)
Dist_Mat = [Dist_MAT_MS50_Copy, Dist_MAT_MS55_Copy, Dist_MAT_MS60_Copy, 
            Dist_MAT_MS65_Copy, MS_S, OS_S]
S = [s_MS50, s_MS55, s_MS60, s_MS65, MS, OS]
F = np.cumsum(np.ones(10000))-1
freqs = [F, F, F, F, MS_freq, OS_freq]

T = [r'Mushroom Spring, $T=50^oC$', r'Mushroom Spring, $T=55^oC$', r'Mushroom Spring, $T=60^oC$', 
     r'Mushroom Spring, $T=65^oC$','Mushroom Spring Samples','Octopus Spring Samples']
fig, ax = plt.subplots(3,2,figsize = (7.01, 1.5*7.01))
HM = []

cmap = mpl.cm.YlOrBr
bounds = np.arange(-0.02, 0.20, 0.02)
norm = mpl.colors.BoundaryNorm(bounds, cmap.N, extend='neither')


for i in range(0, len(S)):
    s, D, freq = S[i], Dist_Mat[i], freqs[i]
    x,y = int(i/2), i%2
    alpha, beta, gamma = (freq[np.where(s == 'Alpha')[0]], freq[np.where(s == 'Beta')[0]], 
                          freq[np.where(s == 'Gamma')[0]])
      
    im = ax[x][y].imshow(D, cmap  = 'YlOrBr',norm = norm, interpolation = 'none')
    HM.append(im)
    
    if i == 3:
        xticks = list(alpha)+list(gamma)
        labels = [r'$OS\minus A$', r'$\gamma$']
    if i <= 2 or i == 5:
        xticks = list(alpha)+list(beta)
        labels = [r'$OS\minus A$', r'$OS\minus B^{\prime}$']
    if i == 4:
        xticks = list(alpha)+list(beta)+list(gamma)
        labels = [r'$OS\minus A$', r'$OS\minus B^{\prime}$', r'$\gamma$']
        
    for u in range(0, len(xticks)):
        k = xticks[u]
        ax[x][y].axvline(k, color = 'black', linestyle = '--', linewidth = 0.75)
        ax[x][y].axhline(k, color = 'black', linestyle = '--', linewidth = 0.75)
    
    ax[x][y].tick_params(top=True, labeltop=True, bottom=False, labelbottom=False)
    ax[x][y].set_xticks(xticks)
    ax[x][y].set_xticklabels(labels, fontsize = 10, rotation = 0,  ha="center")
    ax[x][y].set_yticks(xticks)
    ax[x][y].set_yticklabels(labels, fontsize = 10)
    ax[x][y].set_title(T[i], fontsize = 12)
    
    if i == 4:
        dx = 15/72.; dy = 0/72. 
        offset = mpl.transforms.ScaledTranslation(dx, dy, fig.dpi_scale_trans)
        for label in ax[x][y].xaxis.get_majorticklabels():
            if label.get_text() == '$OS\\minus A$':
                label.set_transform(label.get_transform() - offset)
            
axcolor = fig.add_axes([0.875,0.0475,0.0095,0.86])
fig.text(0.07, 0.965,'A', size = 14)
fig.text(0.5, 0.965,'B', size = 14)
fig.text(0.07, 0.645,'C', size = 14)
fig.text(0.5, 0.645,'D', size = 14)
fig.text(0.07, 0.30,'E', size = 14)
fig.text(0.5, 0.30,'F', size = 14)
fig.colorbar(HM[5], cax = axcolor)
fig.tight_layout()
fig.subplots_adjust(right = 0.85)
fig.savefig(outpath+'Fig-S12.pdf')
```
